Vector - Has the magnitude and direction it is represented in array or list format

```python
vector=torch.tensor([7,7])
print(vector)
print('The dimension - ',vector.ndim)
print(vector.shape)
```

MATRIX

```python
matrix = torch.tensor([[7,7],
                       [9,10]])
print(matrix)
print('The dimension - ',matrix.ndim)
print(matrix.shape)
print('Getting the values based on index i - ',matrix[1])
print('Getting the values based on index i & j - ',matrix[1][1])
```

Tensor

```python
TENSOR = torch.tensor([[[1,2,3],
                        [5,6,7],
                        [10,11,12]
                         ]])

print(TENSOR)
print('The dimension - ',TENSOR.ndim)
print(TENSOR.shape)
print('Getting the values based on index i - ',TENSOR[0]) # It means it has 1 3 * 3 matrix inside the tensor
print('Getting the values based on index i - ',TENSOR[0][1][2])
```
